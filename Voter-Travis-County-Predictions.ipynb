### Get Party Voters

```python
party_columns = [col for col in df_data.columns if "PARTY" in col]
```

```python
party_df = pd.get_dummies(df_data[party_columns])
```

```python
dem_columns = [col for col in party_df.columns if "_DEMO" in col]
```

```python
rep_columns = [col for col in party_df.columns if "_REPB" in col]
```

```python
df_data['DEM_VOTES'] = party_df[dem_columns].T.sum()
```

```python
df_data['REP_VOTES'] = party_df[rep_columns].T.sum()
```
